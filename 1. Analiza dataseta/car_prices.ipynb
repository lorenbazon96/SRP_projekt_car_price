```python
## Opis skupa podataka

Skup podataka sadrži informacije o prodaji rabljenih vozila, uključujući godinu proizvodnje, marku, model, stanje vozila, kilometražu, cijenu i datum prodaje. Cilj analize je dobiti osnovni uvid u strukturu podataka i pripremiti ih za daljnju analizu.
```

```python
import pandas as pd
import os

base_dir = os.getcwd()
csv_path = os.path.join(base_dir, "data", "car_prices.csv")

data = pd.read_csv(csv_path)
data.head()
```

```python
## Veličina skupa podataka
```

```python
print(f"Redovi: {data.shape[0]}")
print(f"Stupci: {data.shape[1]}")
```

```python
## Nazivi stupaca
```

```python
for i, col in enumerate(data.columns, 1):
    print(f"{i}. {col}")
```

```python
## Tipovi podataka
```

```python
data.dtypes
```

```python
## Nedostajuće vrijednosti
```

```python
missing = data.isna().sum()
missing[missing > 0]
```

```python
## Jedinstvene vrijednosti kategorijskih stupaca
```

```python
tekstualni_stupci = [
    'year','make','model','trim','body','transmission','vin','state',
    'condition','odometer','color','interior','seller','mmr',
    'sellingprice','saledate'
]

for col in tekstualni_stupci:
    if col in data.columns:
        unique_vals = data[col].dropna().unique()
        print(f"\n{col}: {len(unique_vals)} jedinstvenih vrijednosti")
        print(unique_vals[:5])
```

```python
## Frekvencije vrijednosti po stupcima
```

```python
for col in tekstualni_stupci:
    if col in data.columns:
        print(f"\n{col}")
        print(data[col].value_counts().head(10))
```
